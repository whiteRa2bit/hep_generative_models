```python
import matplotlib.pyplot as plt
import scipy
import numpy as np
from IPython import display

from generation.config import SIGNALS_TRAINING_CONFIG as config
from generation.dataset.signals_dataset import SignalsDataset
```

```python
dataset = SignalsDataset(signal_dim=config['x_dim'])
noises = dataset.noises
```

```python
SAMPLE_RATE = 1

def plot_noise(noise):
    f, ax = plt.subplots(1, 2, figsize=(12, 5))

    ax[0].set_title("Noise")
    ax[0].plot(noise)
    ax[0].set_xlabel('Sample')
    ax[0].set_ylabel('Amplitude')

    ax[1].set_title("Spectrogram")
    ax[1].specgram(noise, Fs=SAMPLE_RATE)
    ax[1].set_xlabel('Time')
    ax[1].set_ylabel('Frequency')
    
    plt.show()

for noise in noises[0][:100]:
    display.clear_output(wait=True)
    plot_noise(noise)
```

```python
fig = plt.figure(num=None, figsize=(20, 20), dpi=100)
plt.axis('off')
plt.specgram(noise, Fs=SAMPLE_RATE)
fig.savefig("sample.png", bbox_inches='tight', transparent=True, pad_inches=0)
plt.close()
```
